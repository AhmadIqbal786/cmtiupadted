```python
from pathlib import Path
import re
root = Path('..').resolve()
pattern = re.compile(r'^([ \t]*)backdrop-filter: blur\((5|10|12|15)px\);')
modified = []
for path in root.rglob('*'):
    if path.suffix.lower() in {'.html', '.css'}:
        text = path.read_text(encoding='utf-8')
        lines = text.splitlines(True)
        out = []
        changed = False
        for idx, line in enumerate(lines):
            match = pattern.match(line)
            if match:
                indent, px = match.groups()
                prev = out[-1] if out else ''
                expected = f"{indent}-webkit-backdrop-filter: blur({px}px);"
                if not prev.rstrip('\r\n') == expected:
                    out.append(expected + "\n")
                    changed = True
            out.append(line)
        new_text = ''.join(out)
        if changed and new_text != text:
            path.write_text(new_text, encoding='utf-8')
            modified.append(str(path))

print(f"Modified {len(modified)} files")
for p in modified:
    print(p)
```

```python
from pathlib import Path
import re

root = Path('..').resolve()
style_map = {}
next_id_ref = [1]

def normalize_style(style_text):
    props = [p.strip() for p in style_text.replace('\n', ' ').replace('\r', ' ').split(';') if p.strip()]
    return '; '.join(props) + (';' if props else '')

style_attr_pattern = re.compile(r'(<[^>]*?)\sstyle\s*=\s*"([^\"]*?)"([^>]*>)', re.IGNORECASE | re.DOTALL)
class_pattern = re.compile(r'\bclass\s*=\s*"([^\"]*?)"', re.IGNORECASE)

modified_files = []
for html_path in root.rglob('*.html'):
    text = html_path.read_text(encoding='utf-8')
    def replacer(match):
        nonlocal next_id
        before, raw_style, after = match.groups()
        norm = normalize_style(raw_style)
        if not norm:
            return before + after
        cls = style_map.get(norm)
        if not cls:
            cls = f'inline-style-{next_id}'
            next_id += 1
            style_map[norm] = cls
        tag = before + after
        class_match = class_pattern.search(tag)
        if class_match:
            existing = class_match.group(1)
            new_classes = existing + ' ' + cls
            return tag[:class_match.start(1)] + new_classes + tag[class_match.end(1):]
        else:
            return before + f' class="{cls}"' + after
    new_text = style_attr_pattern.sub(replacer, text)
    if new_text != text:
        html_path.write_text(new_text, encoding='utf-8')
        modified_files.append(str(html_path))

css_path = root / 'style.css'
css_text = css_path.read_text(encoding='utf-8')
insert_text = '\n\n/* Auto-generated classes for removed inline styles */\n'
for style, cls in style_map.items():
    insert_text += f'.{cls} {{ {style} }}\n'
if insert_text.strip() != '':
    css_path.write_text(css_text + insert_text, encoding='utf-8')

print(f"Converted {len(style_map)} unique inline styles into classes in {len(modified_files)} HTML files")
for p in modified_files[:50]:
    print(p)
```

```python
from pathlib import Path
from collections import Counter
import re
root = Path('..').resolve()
style_values = Counter()
for html in root.rglob('*.html'):
    text = html.read_text(encoding='utf-8')
    for val in re.findall(r'style="([^"]*)"', text):
        style_values[val.strip()] += 1
for val, count in style_values.most_common():
    print(count, repr(val))
```
