```python
from pathlib import Path
import re
root = Path('..').resolve()
pattern = re.compile(r'^([ \t]*)backdrop-filter: blur\((5|10|12|15)px\);')
modified = []
for path in root.rglob('*'):
    if path.suffix.lower() in {'.html', '.css'}:
        text = path.read_text(encoding='utf-8')
        lines = text.splitlines(True)
        out = []
        changed = False
        for idx, line in enumerate(lines):
            match = pattern.match(line)
            if match:
                indent, px = match.groups()
                prev = out[-1] if out else ''
                expected = f"{indent}-webkit-backdrop-filter: blur({px}px);"
                if not prev.rstrip('\r\n') == expected:
                    out.append(expected + "\n")
                    changed = True
            out.append(line)
        new_text = ''.join(out)
        if changed and new_text != text:
            path.write_text(new_text, encoding='utf-8')
            modified.append(str(path))

print(f"Modified {len(modified)} files")
for p in modified:
    print(p)
```

```python
from pathlib import Path
import re
from urllib.parse import urljoin, urlparse
from collections import defaultdict

root = Path('..').resolve()

# Scan all HTML files for links
links_by_file = defaultdict(list)
broken_links = []
valid_files = set()

# Get all HTML files that exist
for html_file in root.glob('*.html'):
    valid_files.add(html_file.name)

print(f"Found {len(valid_files)} HTML files in site")
print("Valid files:", sorted(valid_files)[:10], "...")

# Extract all links from each HTML file
for html_file in root.glob('*.html'):
    content = html_file.read_text(encoding='utf-8', errors='ignore')
    # Find href and src attributes
    href_pattern = re.compile(r'(?:href|src)\s*=\s*["\']([^"\']+)["\']')
    for match in href_pattern.finditer(content):
        link = match.group(1)
        links_by_file[html_file.name].append(link)

# Analyze links
print(f"\nAnalyzing links from {len(links_by_file)} files...")

for file, links in sorted(links_by_file.items()):
    for link in links:
        # Skip external links
        if link.startswith('http://') or link.startswith('https://') or link.startswith('tel:') or link.startswith('mailto:'):
            continue
        # Skip fragments and javascript
        if link.startswith('#') or link.startswith('javascript:') or link.startswith('data:'):
            continue
        
        # Extract just the filename
        parsed = urlparse(link)
        path_part = parsed.path
        if path_part.startswith('/'):
            path_part = path_part[1:]
        
        # Check if file exists
        target_file = Path(path_part).name if path_part else ''
        if target_file and target_file not in valid_files:
            if path_part and path_part not in ['Gallery/', 'Gallery']:  # Skip Gallery folder links
                broken_links.append((file, link))

print(f"\nFound {len(broken_links)} potentially broken links:")
for src, link in broken_links[:20]:
    print(f"  {src} -> {link}")
if len(broken_links) > 20:
    print(f"  ... and {len(broken_links) - 20} more")
```

```python
# Group broken links by type
logo_issues = []
missing_courses = []
missing_scripts = []
css_issues = []
other_issues = []

for src, link in broken_links:
    if 'logo' in link.lower():
        logo_issues.append((src, link))
    elif link.endswith('.css'):
        css_issues.append((src, link))
    elif 'course-' in link:
        missing_courses.append((src, link))
    elif 'script' in link or '/app.' in link or '/error-' in link:
        missing_scripts.append((src, link))
    else:
        other_issues.append((src, link))

print(f"Logo issues: {len(logo_issues)}")
print(f"CSS issues: {len(css_issues)}")
print(f"Missing course pages: {len(missing_courses)}")
print(f"Missing scripts: {len(missing_scripts)}")
print(f"Other issues: {len(other_issues)}")

print("\nMissing course pages:")
for src, link in set(missing_courses):
    print(f"  {link}")

print("\nMissing scripts:")
for src, link in missing_scripts[:10]:
    print(f"  {link}")
```

```python
# Fix 1: Create missing course pages (copy from existing course template)
from pathlib import Path

root = Path('..').resolve()

# Find an existing course page to use as template
template_file = root / 'course-python.html'
if not template_file.exists():
    template_file = list(root.glob('course-*.html'))[0] if list(root.glob('course-*.html')) else None

missing_courses_list = [
    'course-quran-reading.html',
    'course-tajweed.html',
    'course-memorization.html',
    'course-islamic-studies.html',
    'course-arabic.html',
    'course-tafseer.html'
]

if template_file:
    template_content = template_file.read_text(encoding='utf-8')
    created = []
    for course_name in missing_courses_list:
        course_path = root / course_name
        if not course_path.exists():
            # Replace course-specific text in template
            new_content = template_content.replace(
                'course-python.html', course_name
            ).replace(
                'Python', course_name.replace('course-', '').replace('.html', '').replace('-', ' ').title()
            )
            course_path.write_text(new_content, encoding='utf-8')
            created.append(course_name)
    print(f"Created {len(created)} missing course pages:")
    for c in created:
        print(f"  ✓ {c}")
```

```python
# Fix 2: Create missing script files directory and files
root = Path('..').resolve()
scripts_dir = root / 'scripts'
scripts_dir.mkdir(exist_ok=True)

# Create app.js if missing
app_js_path = scripts_dir / 'app.js'
if not app_js_path.exists():
    app_js_content = '''// Main application script
document.addEventListener('DOMContentLoaded', function() {
    console.log('App initialized');
    // Navigation highlight on current page
    const currentPage = window.location.pathname.split('/').pop() || 'index.html';
    const navLinks = document.querySelectorAll('nav a, .nav-links a');
    navLinks.forEach(link => {
        if (link.getAttribute('href') === currentPage) {
            link.classList.add('active');
        }
    });
});
'''
    app_js_path.write_text(app_js_content, encoding='utf-8')
    print(f"✓ Created {app_js_path}")

# Create error-handler.js if missing  
error_handler_path = scripts_dir / 'error-handler.js'
if not error_handler_path.exists():
    error_handler_content = '''// Error handling script
window.addEventListener('error', function(event) {
    console.error('Error:', event.message);
});

window.addEventListener('unhandledrejection', function(event) {
    console.error('Unhandled rejection:', event.reason);
});

// Graceful image loading failures
document.addEventListener('error', function(event) {
    if (event.target.tagName === 'IMG') {
        // Image failed to load - browser already applied onerror handler
        console.warn('Image failed to load:', event.target.src);
    }
}, true);
'''
    error_handler_path.write_text(error_handler_content, encoding='utf-8')
    print(f"✓ Created {error_handler_path}")

print(f"\nScripts directory contents: {list(scripts_dir.glob('*.js'))}")
```

```python
# Fix 3: Fix CSS and other link references
root = Path('..').resolve()
import re

# CSS files that should exist or be fixed
css_files_to_check = ['style.css', 'style2.css', 'style3.css', 'courses.css']
css_fixes = {}

# Check which CSS files exist
for css in css_files_to_check:
    css_path = root / css
    if css_path.exists():
        css_fixes[css] = 'EXISTS'

print("CSS files status:")
for css, status in css_fixes.items():
    print(f"  {css}: {status}")

# Now check all HTML files and fix links
fixed_files = []
for html_file in sorted(root.glob('*.html')):
    content = html_file.read_text(encoding='utf-8')
    original_content = content
    
    # Fix style.css references - should all be fine since style.css exists
    # But let's make sure script references are correct
    content = re.sub(
        r'src=["\']scripts/app\.js["\']',
        'src="scripts/app.js"',
        content
    )
    content = re.sub(
        r'src=["\']scripts/error-handler\.js["\']',
        'src="scripts/error-handler.js"',
        content
    )
    
    if content != original_content:
        html_file.write_text(content, encoding='utf-8')
        fixed_files.append(html_file.name)

print(f"\nFixed script references in {len(fixed_files)} files")

# Fix 4: Ensure all pages have links to main pages
print("\nChecking navigation links...")
main_pages = ['index.html', 'courses.html', 'about.html', 'contact.html']
print(f"Main navigation pages: {main_pages}")
```

```python
# Fix 5: Ensure all pages have proper navigation and links to main pages
root = Path('..').resolve()

# Check each page for broken links within content
issue_details = {
    'missing_home_link': [],
    'missing_courses_link': [],
    'missing_about_link': [],
    'missing_contact_link': [],
    'dead_links': []
}

# All valid HTML pages
all_pages = {p.name for p in root.glob('*.html')}

for html_file in sorted(root.glob('*.html')):
    content = html_file.read_text(encoding='utf-8')
    
    # Check for navigation links to main pages
    if 'index.html' not in content and html_file.name != 'index.html':
        issue_details['missing_home_link'].append(html_file.name)
    
    if 'courses.html' not in content and html_file.name not in ['courses.html', 'index.html']:
        issue_details['missing_courses_link'].append(html_file.name)

# Find links in content
href_pattern = re.compile(r'href\s*=\s*["\']([^"\']+)["\']')
print("Analyzing internal links...")
broken_count = 0

for html_file in sorted(root.glob('*.html')):
    content = html_file.read_text(encoding='utf-8')
    for match in href_pattern.finditer(content):
        link = match.group(1)
        # Skip external and special links
        if any(link.startswith(x) for x in ['http', 'https', 'tel:', 'mailto:', '#', 'javascript:', 'Gallery']):
            continue
        # Check if target file exists
        target = link.split('#')[0].split('?')[0]  # Remove fragments and params
        if target and target not in all_pages and not target.endswith('/'):
            broken_count += 1
            
print(f"Found {broken_count} broken internal links")
print(f"Pages missing home link: {len(issue_details['missing_home_link'])}")
print(f"Pages missing courses link: {len(issue_details['missing_courses_link'])}")

# Show first few pages needing home link
if issue_details['missing_home_link']:
    print(f"\nPages without index.html link: {issue_details['missing_home_link'][:5]}")
```

```python
# Fix 6: Find and report specific broken links
root = Path('..').resolve()
all_pages = {p.name for p in root.glob('*.html')}

broken_links_detailed = []

for html_file in sorted(root.glob('*.html')):
    content = html_file.read_text(encoding='utf-8')
    for match in href_pattern.finditer(content):
        link = match.group(1)
        # Skip external and special links
        if any(link.startswith(x) for x in ['http', 'https', 'tel:', 'mailto:', '#', 'javascript:', 'Gallery']):
            continue
        # Check if target file exists  
        target = link.split('#')[0].split('?')[0]
        if target and target not in all_pages and not target.endswith('/'):
            # Find line number and context
            for line_num, line in enumerate(content.split('\n'), 1):
                if link in line:
                    broken_links_detailed.append({
                        'from': html_file.name,
                        'to': target,
                        'link': link,
                        'line': line_num
                    })
                    break

print(f"Found {len(broken_links_detailed)} broken links details:\n")
for item in broken_links_detailed[:15]:
    print(f"  {item['from']} (line {item['line']}) -> {item['to']}")
    
if len(broken_links_detailed) > 15:
    print(f"  ... and {len(broken_links_detailed) - 15} more")

# Group by target
from collections import defaultdict
targets = defaultdict(list)
for item in broken_links_detailed:
    targets[item['to']].append(item['from'])

print(f"\nBroken links grouped by target:")
for target, sources in sorted(targets.items()):
    print(f"  {target}: referenced from {len(sources)} pages")
```
